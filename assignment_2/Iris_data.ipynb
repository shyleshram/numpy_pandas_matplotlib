```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris['data']
```

```python
iris['feature_names']
```

```python
iris['target']
```

```python
iris['target_names']
```

```python
iris_data = pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
iris_data.head()
```

```python
target_names = pd.Series(iris['target'])
target_names
```

```python
iris_data['target_names'] = target_names
```

```python
iris_data
```

```python
plt.style.use('seaborn')
plt.scatter(data=iris_data,x='sepal length (cm)',y='sepal width (cm)',
           c='green')
```

```python
sns.jointplot(data=iris_data,x='sepal length (cm)',y='sepal width (cm)',
             kind='hex')
```

# CLEARY WE CAN THREE CLUSTER

```python
sns.pairplot(iris_data, hue='target_names')
```

```python
sns.pairplot(x_vars='sepal length (cm)',y_vars='sepal width (cm)',data=iris_data,hue='target_names',size=7)
```

# 'SETOSA' can be classified easily based on these two features. It stands out. But 'versicolor', and  'virginica' have 
some similarities at mid-range, But 'virginica' speal length is higer than 'versicolor' species.
